```python
#accessing col
import pandas as pd
data = [
    ['kir', 20],
    ['ramesh', 30],
    ['suresh', 40]
]
# Create DataFrame with column names
df = pd.DataFrame(data, columns=['name', 'age'], index=[100, 101, 102])
print(type(df['name']))
print("-----------------------------------")

print(df[['name']]) #now dataframe
print("-----------------------------------")

#new
df['salary']=[1000,20000,3000]
df['bonus']=df['salary']*0.10
print(df)
print("-----------------------------------")

#to delete a col
del df['bonus']
print(df)
```

```python
#accessing row

import pandas as pd
data = {
    "name": ["Asha", "Ravi", "Kiran", "Meena"],
    "age": [22, 25, 23, 28],
    "salary": [30000, 45000, 40000, 50000],
    "department": ["IT", "HR", "IT", "Finance"]
}
df = pd.DataFrame(data)
print(df.loc[0])
print("----------------------------------------")

print(df.iloc[1])
print("----------------------------------------")

print(df.loc[[0,1]])
print("----------------------------------------")

print(df[0:3])

```

```python
import pandas as pd
data = {
    "name": ["Asha", "Ravi", "Kiran", "Meena"],
    "age": [22, 25, 23, 28],
    "salary": [30000, 45000, 40000, 50000],
    "department": ["IT", "HR", "IT", "Finance"]
}
df = pd.DataFrame(data)

print(df.head()) #print first 5
print(df.tail()) #print last 5
print(df.shape) #(rows,cols)
print("------------------")

print(df.columns)
print("------------------")

print(df.info())
print("-------------------------")

df.describe()

```

```python
#filtering

import pandas as pd
import numpy as np

data = {
    "name": ["Asha", "Ravi", "Kiran", "Meena", "Suresh"],
    "age": [22, 25, 23, 28, 30],
    "salary": [30000, 45000, 40000, 50000, 60000],
    "department": ["IT", "HR", "IT", "Finance", "IT"]
}

df = pd.DataFrame(data)
print(df[df["age"] > 25])
print("-----------------------------------")

print(df[(df["age"] > 25) & (df["salary"] > 50000)])
print("-----------------------------------")


print(df[(df["age"] < 23) | (df["department"] == "IT")])


```

```python
#handling missing values
data = {
    "name": ["Asha", "Ravi", "Kiran", "Meena"],
    "age": [22, np.nan, 23, 28],
    "salary": [30000, 45000, np.nan, 50000]
}

df = pd.DataFrame(data)
print(df.isnull())
print("-----------------------------------")

print(df.dropna()) #removes rows with missing data
print("-----------------------------------")

print(df.fillna(0)) #fill missing data with 0
print("-----------------------------------")
```
