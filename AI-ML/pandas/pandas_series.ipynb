```python
#series ->1d labeled array
#create a series from list,numpy arrays,dictionaries
#indexing & slicing
#vectorized operations
# label vs positional indexing

import pandas as pd
#list to series
data = [10,20,30,40]  #no label specified so 0 1 2 (label=index)
s=pd.Series(data)
print(s)

data = [10,20,30,40]  #list to series, label specified so 11 22 33 44 (label=index)
s=pd.Series(data,[11,22,33,44])
print(s)


#numpy arr to series
import numpy as np
arr=np.array([11,12,13,14])
s=pd.Series(arr,index = [11,22,33,44])
print(s)

#dict to series
d={'a':100,'b':200,'c':300}
s=pd.Series(d)
print(s)
```

```python
#indexing & slicing
import pandas as pd

data = [10,20,30,40]
s=pd.Series(data,index=[1,2,3,4])
print(s[2])

print(data[0:3]) #slicing in list
print(s[0:3]) #slicing in series


print(s.values)
print(s.index)
print(s.dtype)
print(s.shape)


```

```python
#vectorized operations
data = [10,20,30,40]
s=pd.Series(data,index=['a','b','c','d'])
print(s)
print(s*2)
print(data * 2) #list la we cant do vectorized operations, it adds ele 2 times,only looping works

data1 = [100,200,300,400]
s1=pd.Series(data1,index=['a','b','c','d'])
data2 = [100,200,300,400]
s2=pd.Series(data2,index=['a','b','c','d'])
print(s1+s2)

#filtering data
marks = pd.Series([70, 80, 90])
print(marks + 5)
print(marks[marks > 75])
```

```python
import pandas as pd
data = [100,200,300,400]
s1=pd.Series(data,index=['a','b','c','d'])
print(s1['a'])
print(s1.iloc[0]) #position based
print(s1.loc['a']) #label based
```

```python
import pandas as pd
s = pd.Series([10, 20, 20, 30])

print(s.sum())
print(s.mean())
print(s.max())
print(s.value_counts())
```

```python
s = pd.Series([10, None, 30])
print(s)
print(s.isnull())
print(s.fillna(0)) #fill missing as 0
```
